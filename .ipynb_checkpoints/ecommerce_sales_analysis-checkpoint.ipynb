```python
import pandas as pd

df = pd.read_csv("ecommerce_sales_dataset_with_sales_and_profit.csv")

df.head()

```

```python
df['order_date']=pd.to_datetime(df['order_date'])
```

```python
df['month'] = df['order_date'].dt.to_period('M')
```

```python
df.duplicated().sum()
```

```python
total_sales = df['total_price'].sum()
total_sales
```

```python
total_profit = df['profit'].sum()
total_profit
```

```python
total_orders = df['order_id'].nunique()
total_orders
```

```python
avg_order_value = total_sales / total_orders
avg_order_value
```

```python
import matplotlib.pyplot as plt
```

```python
monthly_sales = df.groupby('month')['total_price'].sum()

monthly_sales.plot(kind='line', title='Monthly Sales Trend', figsize=(8,5))
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_categories = df.groupby('product_category')['total_price'].sum().sort_values(ascending=False).head(5)

top_categories.plot(kind='bar', title='Top 5 Categories by Sales', figsize=(8,5))
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
profit_category = df.groupby('product_category')['profit'].sum().sort_values(ascending=False)

profit_category.plot(kind='bar', title='Profit by Category', figsize=(8,5))
plt.show()
```
